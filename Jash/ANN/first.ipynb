```python
import pandas as pd
import os

current = os.getcwd()
root = os.path.join(current, "..")
data_folder = os.path.join(root, "WDN1", "csv-dist-orig")
labelled_folder = os.path.join(root, "WDN1", "csv-updated")

files = os.listdir(data_folder)
files.pop()

def label_csv(file):
    
    name = file.split(".")[0]

    data = pd.read_csv(os.path.join(data_folder, file), header=None)
    column_names = []

    column_names.append("time_seconds")

    for i in range(2, 10):
        column_names.append(f"flow_node{i}")
        
    for i in range(2, 10):
        column_names.append(f"pressure_node{i}")
        
    for i in range(2, 10):
        column_names.append(f"volume_node{i}")
        
    for i in range(2, 10):
        column_names.append(f"x_node{i}")
        
    for i in range(2, 10):
        column_names.append(f"y_node{i}")
        
    for i in range(2, 10):
        column_names.append(f"z_node{i}")
        
    for i in range(10, 15):
        column_names.append(f"leak_node{i}")
        
    for coord in ["x", "y", "z"]:
        column_names.append(f"leak_{coord}")
        
    column_names.append("weekday_flag")

    print("Total columns: ", len(column_names))
    print("Data shape: ", data.shape)

    data.columns = column_names
    data.to_csv(os.path.join(labelled_folder, f"labelled_{name}.csv"), index=False)
    print(f"Updated {name} to folder and saved.")
    
for file in files:
    label_csv(file)
```
