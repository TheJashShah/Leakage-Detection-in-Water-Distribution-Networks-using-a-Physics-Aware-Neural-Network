```python
import networkx as nx
import matplotlib.pyplot as plt

class Graph:
    
    def __init__(self):
        self.list = []
        
    def addTemp(self, a, b):
        temp = [a, b]
        self.list.append(temp)
        
    def visualize(self):
        G = nx.Graph()
        G.add_edges_from(self.list)
        nx.draw_networkx(G)
        plt.show()
        
G = Graph()

for i in range(len(node1)):
    G.addTemp(node1[i], node2[i])
    
G.visualize()
    
```
