```python
import os
import pandas as pd
import glob

current = os.getcwd()
root = os.path.join(current, "..")
labelled_folder = os.path.join(root, "WDN1", "csv-updated")

file_list = os.listdir(labelled_folder)
all_data = []

for file in file_list:
    day = file.split(".")[0]
    data = pd.read_csv(os.path.join(labelled_folder, file))
    
    leak_node = None
    for i in range(10, 15):
        if data[f"leak_node{i}"].sum() > 0:
            leak_node = i
            break
        
    if leak_node is None:
        data["leak_status"] = 0
    else:
        data["leak_status"] = leak_node
        
    data["is_leak"] = (data["leak_status"] > 0).astype(int)
    data["day_id"] = day
    
    all_data.append(data)
    
final_dataframe = pd.concat(all_data, ignore_index=True)
final_dataframe.head(20)

final_dataframe.to_csv(os.path.join(current, "merged.csv"), index=False)
```
