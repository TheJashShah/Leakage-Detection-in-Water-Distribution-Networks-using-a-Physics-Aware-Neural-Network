```python
import os

current = os.getcwd()
Leakdb = os.path.join(current, "..")
csv_dir = os.path.join(Leakdb, "csv_of_inp")
```

```python
files = os.listdir(csv_dir)

import pandas as pd

dfs = []

for file in files:
    path = os.path.join(csv_dir, file)
    df = pd.read_csv(path)
    dfs.append(df[["pipe_id", "length", "diameter", "roughness"]])
    
combined = pd.concat(dfs)

average_df = combined.groupby("pipe_id", as_index=False).mean()

average_df = average_df.rename(columns={
    'length': 'average_length',
    'diameter': 'average_diameter',
    'roughness': 'average_roughness'
})

average_df.to_csv(os.path.join(Leakdb, "average_values.csv"), index=False)
merged_df = combined.merge(average_df, on='pipe_id')

merged_df['length_deviation'] = merged_df['length'] - merged_df['average_length']
merged_df['diameter_deviation'] = merged_df['diameter'] - merged_df['average_diameter']
merged_df['roughness_deviation'] = merged_df['roughness'] - merged_df['average_roughness']

merged_df['length_deviation_%'] = (merged_df['length_deviation'] / merged_df['average_length']) * 100
merged_df['diamter_deviation_%'] = (merged_df['diameter_deviation'] / merged_df['average_diameter']) * 100
merged_df['roughness_deviation_%'] = (merged_df['roughness_deviation'] / merged_df['average_roughness']) * 100

merged_df.to_csv(os.path.join(Leakdb, "deviations.csv"), index=False)
```

```python
deviation_summary = merged_df.groupby('pipe_id').agg({
    'length_deviation_%': ['mean', 'mean', lambda x: x.abs().mean()],
    'diamter_deviation_%': ['mean', lambda x: x.abs().mean()],
    'roughness_deviation_%': ['mean', lambda x: x.abs().mean()],
})
```
